```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cPickle as pickle
%matplotlib inline
```

```python
def loadInfo(fileName):
    with open(fileName,'r') as fid:
        opts = pickle.load(fid)
        costs = pickle.load(fid)
    return opts,costs

def smooth(costs,winSize=20):
    """
    Smooth costs using windowed average
    """
    newC = np.convolve(costs,np.ones((winSize))/float(winSize),'valid')
    return newC

def loadAll(files):
    cList = []
    for f in files:
        _,costs = loadInfo(f)
        newC = smooth(costs)
        cList.append(newC)
    return np.vstack(cList)
```

```python
files = []
for step in [1e-1,1e-2,1e-3]:
    files.append("../models/sgd_step_"+str(step)+"_mom_"+str(1)+".bin")
costMat = loadAll(files)
print "All sgd no annealing"
plt.plot(costMat.T)
plt.show()

files = []
for step in [1e-1]:
    files.append("../models/sgd_step_"+str(step)+"_mom_"+str(1)+".bin")
costMat = loadAll(files)
print "SGD High LR"
plt.plot(costMat.T[100:,:])
plt.show()
```
