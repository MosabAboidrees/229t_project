```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cPickle as pickle
%matplotlib inline
```

```python
def loadInfo(fileName):
    with open(fileName,'r') as fid:
        opts = pickle.load(fid)
        costs = pickle.load(fid)
    return opts,costs

def smooth(costs,winSize=20):
    """
    Smooth costs using windowed average
    """
    newC = np.convolve(costs,np.ones((winSize))/float(winSize),'valid')
    return newC

def loadAll(files):
    cList = []
    for f in files:
        _,costs = loadInfo(f)
        newC = smooth(costs)
        cList.append(newC)
    return np.vstack(cList)
```

```python
files = []
formatf = "../models/ae_%s_step_%.3f_mom_%.3f_anneal_%.3f.bin"
files.append(formatf%('nesterov',1e-2,0.99,1.0))
files.append(formatf%('adagrad',5e-3,1.0,1.0))
files.append(formatf%('adagrad3',5e-3,1.0,1.0))
files.append(formatf%('adadelta',5e-3,1.0,1.0))
costMat = loadAll(files)
print "Autoencoder All comparison"
plt.plot(costMat.T[500:,:])
plt.show()
```
